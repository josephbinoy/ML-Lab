```python
import pandas as pd

df = pd.read_csv('CarPrice_Assignment.csv')
df
```

```python
print(df.isna().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df['fueltype']=le.fit_transform(df['fueltype'])
df
```

```python
features = ['horsepower', 'enginesize']
X=df[features]
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2, max_iter=300, n_init=10)
model.fit(X)
```

```python
centroids = model.cluster_centers_
centroids
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['horsepower'], df['enginesize'])
for c in centroids:
    plt.scatter(c[0], c[1], c='r')
    
plt.show()
```

```python
clusters = model.predict(X)
```

```python
df['Cluster'] = clusters
```

```python
import numpy as np

for cluster in clusters:
    plt.scatter(df[df['Cluster'] == cluster]['enginesize'],df[df['Cluster'] == cluster]['horsepower'])
```

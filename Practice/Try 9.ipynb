```python
from sklearn.datasets import load_iris

dataset = load_iris()
X=dataset.data
y=dataset.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_scaled=scaler.fit_transform(X)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

#Perform AGNES Algorithm
Z1 = linkage(X_scaled,method='single')

#Plot Dendogram
dendrogram(Z1)
plt.title('AGNES Dendrogram')
plt.xlabel("Samples")
plt.ylabel("Distances")
plt.show()
```

```python
#Perform DIANA Algorithm
Z2 = linkage(X_scaled,method='average')

dendrogram(Z2)
plt.title('DIANA Dendrogram')
plt.xlabel('samples')
plt.ylabel('distance')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

agnus_clusters=fcluster(Z1, 0.8 ,criterion='distance')

plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1],c=agnus_clusters)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()
plt.show()
```

```python
diana_clusters=fcluster(Z2,0.8,criterion='distance')

plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1],c=diana_clusters)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()
plt.show()
```
